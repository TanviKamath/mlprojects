```python
numeric_features = [feature for feature in df.columns if df[feature].dtype!='O']
categorical_features = [feature for feature in df.columns if df[feature].dtype=='O']
print("Numeric Features:", numeric_features)
print("Categorical Features:", categorical_features)
```

```python
df.head()
```

```python
df['total_marks'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average_marks'] = df['total_marks'] / 3
```

```python
df.head()
```

```python
math_full = df[df['math_score']==100]['average_marks'].count()
print("Number of students who scored full marks in Math:", math_full)
reading_full = df[df['reading_score']==100]['average_marks'].count()
print("Number of students who scored full marks in Reading:", reading_full)
writing_full = df[df['writing_score']==100]['average_marks'].count()
print("Number of students who scored full marks in Writing:", writing_full)
```

```python
math_less_20 = df[df['math_score']<20]['average_marks'].count()
print("Number of students who scored less than 20 in Math:", math_less_20)
reading_less_20 = df[df['reading_score']<20]['average_marks'].count()
print("Number of students who scored less than 20 in Reading:", reading_less_20)
writing_less_20 = df[df['writing_score']<20]['average_marks'].count()
print("Number of students who scored less than 20 in Writing:", writing_less_20)
```

```python
sns.histplot(data=df, x='average_marks', kde=True, hue='gender')
plt.title('Average Total Marks by Gender')
plt.ylabel('Average Total Marks')
plt.show()
```

```python
sns.histplot(data=df, x='total_marks', kde=True, hue='gender')
plt.title('Average Total Marks by Gender')
plt.ylabel('Average Total Marks')
plt.show()
```
